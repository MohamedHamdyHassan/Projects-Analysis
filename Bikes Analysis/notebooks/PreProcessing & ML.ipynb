```python
import numpy as np
import pandas as pd
from datasist.structdata import detect_outliers
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import joblib
```

```python
raw_df=pd.read_csv('../data/processed/Analysis & Viz.csv')
df=raw_df.copy()
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0','casual','registered','datetime'],axis=1,inplace=True)
```

```python
df.shape
```

```python
df
```

```python
numeric_data=df.select_dtypes(include='number')
numeric_data.head()
```

```python
numeric_data_col=numeric_data.columns
```

```python
len(detect_outliers(df,0,numeric_data_col))
```

```python
for col in numeric_data_col:
    idx=detect_outliers(df,0,[col])
    df[col].iloc[idx]=df[col].median()
```

```python
len(detect_outliers(df,0,numeric_data_col))
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df=pd.get_dummies(df,columns=['season','weather','day'],drop_first=True)
```

```python
df.columns
```

```python
df.head()
```

```python
import pandas as pd

def replace_bool_with_numbers(df):
    df.replace({False: 0, True: 1}, inplace=True)


df.apply(replace_bool_with_numbers)

```

```python
df.head()
```

```python
x=df.drop(['profit'],axis=1)
y=df['profit']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train_scaled,y_train)
```

```python
model.score(x_train_scaled,y_train)
```

```python
model.score(x_test_scaled,y_test)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
joblib.dump(scaler,'scaler.h5')
```

```python
joblib.dump(model,'model.h5')
```
